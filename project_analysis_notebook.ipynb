# Project 1: Global Consumer Spending Analysis (2010–2024)
This notebook analyzes global consumer spending data across major regions between 2010 and 2024.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'project1_consumer_spending_2010_2024.csv'
df = pd.read_csv(file_path)
df.head()
```

## Summary Statistics

```python
df.describe()
```

## Total Consumer Spending by Region

```python
region_summary = df.groupby('Region')['Consumer_Spending_Billion_USD'].sum().sort_values(ascending=False)
region_summary.plot(kind='bar', title='Total Consumer Spending by Region (2010–2024)', figsize=(8,5))
plt.ylabel('Spending (Billion USD)')
plt.show()
```

## Spending Trends Over Time

```python
for region in df['Region'].unique():
    subset = df[df['Region'] == region]
    plt.plot(subset['Year'], subset['Consumer_Spending_Billion_USD'], label=region)
plt.legend()
plt.title('Consumer Spending Trend (2010–2024)')
plt.xlabel('Year')
plt.ylabel('Spending (Billion USD)')
plt.show()
```
